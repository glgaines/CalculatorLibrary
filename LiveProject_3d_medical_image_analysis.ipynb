### investigate the images

```
my_path = '/content/small'
image = nib.load(os.path.join(my_path, 't1/IXI102-HH-1416-T1_fcm.nii.gz'))
image_data = image.get_fdata()
image_data.shape


images = ['t1/IXI102-HH-1416-T1_fcm.nii.gz', 't2/IXI102-HH-1416-T2_reg_fcm.nii.gz']
slices = [1,20,40,60,80,89]
rows = len(slices)

fig, axes = plt.subplots(nrows = rows, ncols = 2, sharex = True, figsize=(5,20))
for i,image in enumerate(images):
  img = nib.load(os.path.join(my_path, image))
  image_data = img.get_fdata()
  for j,sl in enumerate(slices):
    axes[j,i].imshow(image_data[sl])
```
